```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Find out how many countries are there in the dataset.

```python
df = pd.read_csv('Dataset_w6.csv')
```

```python
print("Number countries are there in the dataset:",df.country.count())
```

# 2.Examine the data by using visualizations

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = df['GDP per capita (current US$)'], y = df['Agricultural production index '],c='green', s=75,alpha=0.5)
plt.xlabel("GDP per capita (current US$)")
plt.ylabel("Agricultural production index ")
plt.title("GDP per capita with respect to Agricultural production index ")
plt.show()
```

```python
data1=df[['country','GDP per capita (current US$)']]
high_gdp=data1.nlargest(5,'GDP per capita (current US$)')
high_gdp.plot(kind='bar',x='country',figsize=(6,5),title='Top 5 GDP Countries',color='cyan')
```

# 3.Find out how many countries with respect to the region are available in the dataset

```python
df['Region'].value_counts()
```

# 4. On average which region has the highest

a. GDP per capita (current US$)

```python
high_dgp = df.groupby('Region')['GDP per capita (current US$)'].mean() 
high_dgp.nlargest(1)
```

b. International trade: Imports (million US$)

```python
high_imports = df.groupby('Region')['International trade: Imports (million US$)'].mean() 
high_imports.nlargest(1)
```

# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
std1 = df.groupby('Region')['GDP per capita (current US$)'].std() 
print("Region has more consistent with respect to GDP percapita (current US$) is:")
std1.nsmallest(1)
```
